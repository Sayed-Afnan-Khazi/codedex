```python
# Importing cool packages
import nltk
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
nltk.download('punkt_tab')
```

```python
# n_value for 'n' grams (>=2 to create bi-grams)
n_value = 5
```

```python
# Reading our training data
fobj = open('./training_data_combined.txt','r')
text = fobj.read().lower()
```

```python
# Tokenizing our training data
tokens  = nltk.word_tokenize(text)
```

```python
tokens
```

```python
total_num_tokens = len(tokens)
total_num_tokens
```

```python
# Forming bi-grams
ngms = list(ngrams(tokens,n_value))
```

```python
ngms
```

```python
# Recording word-pair frequencies
from collections import defaultdict

freq = defaultdict(lambda: defaultdict(int)) # 2D hash table - Essentially a table with rows and columns and a default value of 0 # We're using lambda to create a callable parameter that's required by defaultdict.

for entry in ngms:
    freq[entry[:-1]][entry[-1]] += 1

freq
```

```python
# Store probablilities in `freq` by dividing each entry by the number of tokens
# Optional, but pretty cool
for entry in ngms:
    freq[entry[:-1]][entry[-1]] /= total_num_tokens

freq
```

```python
import heapq
# Create our model using the frequencies
model = {}

for entry in freq.items():
    h = list(entry[-1].items())
    heapq.heapify(h)
    model[entry[0]] = h

model
```

```python
import time, random
output = 'Giant cat is a beautiful' # Must be lowercase
print(output,end='')

tokens_generated = 0
while True:
    # Get the last n-1 "tokens"
    input_tokens = word_tokenize(output)
    last_n_tokens = input_tokens[-(n_value-1):]
    
    # Get best prediction (but make it non-deterministic using random.choice())
    # print(model[tuple(last_n_tokens)])


    if tuple(last_n_tokens) in model:
        next_token = ' ' + random.choice(model[tuple(last_n_tokens)])[0]
    else:
        next_token = ' ' + random.choice(tokens)
    print(next_token, end='')
    output += next_token

    tokens_generated += 1
    if tokens_generated % 10 == 0:
        print() 
    time.sleep(0.1)
```
